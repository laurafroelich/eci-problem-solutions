# Problem 4.15 (ANMs)

## a)
Consider the SCM

\begin{align}
X &:= N_X \\
Y &:= 2X + N_Y,\\
N_X & \sim \mathcal{U}(1, 3) \\
N_Y & \sim \mathcal{U}(-0.5, 0.5),
\end{align}

where $N_X$ and $N_Y$ are independent random variables. 

First, we plot the support of the joint distribution of $X$ and $Y$.

```python
x = np.array([1, 3])
y1 = 2*x - 0.5
y2 = 2*x + 0.5

fig, ax = plt.subplots(1, 1, figsize=(3, 3))

ax.fill_between(x, y1, y2)
ax.set_title('Support of joint distribution of $X$ and $Y$, $P_{X, Y}$.')

ax.set_xlabel('X')
ax.set_ylabel('Y')

fig.tight_layout()
```

We now turn to the question of whether this joint distribution allows an SCM where $X$ is a cause of $Y$, in the form:


\begin{align}
X &:= g(Y) + M_X \\
Y &:= M_Y,\\
\end{align}

where $M_X$ and $M_Y$ are independent random variables.

From the plot above, it is to see that the support of $M_X$ would differ depending on the value of $Y$, and thus of $M_Y$, violating the requirement that  $M_X$ and $M_Y$ are independent. Hence we conclude that there is no ANM from $Y$ to $X$. 

## b)
Consider the SCM

\begin{align}
X &:= N_X \\
Y &:= X^2 + N_Y,\\
N_X & \sim \mathcal{U}(1, 3) \\
N_Y & \sim \mathcal{U}(-0.5, 0.5),
\end{align}

where $N_X$ and $N_Y$ are independent random variables. 

First, we plot the support of the joint distribution of $X$ and $Y$.

```python
x = np.linspace(1, 3, 10)
y1 = x**x - 0.5
y2 = x**x + 0.5

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.fill_between(x, y1, y2)
ax.set_title('Support of joint distribution of $X$ and $Y$, $P_{X, Y}$.')

ax.set_xlabel('X')
ax.set_ylabel('Y')

fig.tight_layout()
```

By the same logic as in a), we see that the support of a noise variable for $X$ cannot be independent of $Y$.

\end{align}

where $X$ and $N_Y$ are independent and the function $f$ is known. 

## a)

Prove that $f(x) = \mathbb{E}[Y|X=x] - \mu_{N_Y}$ (see http://web.math.ku.dk/~peters/jonas_files/errata.pdf).

Let us calculate the quantity $\mathbb{E}[Y|X=x]$:

\begin{align}
\mathbb{E}[Y|X=x] & = \mathbb{E}[f(X) + N_Y|X=x] = \mathbb{E}[f(X)|X=x] + \mathbb{E}[N_Y|X=x] = f(x) + \mathbb{E}[N_Y|X=x] = f(x) + \mathbb{E}[N_Y] = f(x) + \mu_{N_Y} \implies \\
f(x) & = \mathbb{E}[Y|X=x] - \mu_{N_Y}
\end{align}
